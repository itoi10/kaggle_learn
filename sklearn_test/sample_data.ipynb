```python
## サンプルデータ自動生成
# make_blobs 複数の塊に分かれるデータを生成する
from sklearn.datasets import make_blobs
import pandas as pd

X, y = make_blobs(
    random_state=3, # シード値
    n_features=2,   # 特徴量の数
    centers=2,      # 塊の数
    cluster_std=1,  # 標準偏差
    n_samples=300   # データ個数
)

df = pd.DataFrame(X)
df['target'] = y
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# targetの値別にdfを分ける
df0 = df[df['target']==0]
df1 = df[df['target']==1]
# 散布図描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color='b', alpha=0.5)
plt.scatter(df1[0], df1[1], color='r', alpha=0.5)
plt.grid()
plt.show()
```

```python
# 塊が3つの場合のデータ
X, y = make_blobs(
    random_state=3,
    n_features=2,  
    centers=3,      # 塊の数　3
    cluster_std=1, 
    n_samples=300   
)

df = pd.DataFrame(X)
df['target'] = y

# targetの値別にdfを分ける
df0 = df[df['target']==0]
df1 = df[df['target']==1]
df2 = df[df['target']==2]

# 散布図描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color='b', alpha=0.5)
plt.scatter(df1[0], df1[1], color='r', alpha=0.5)
plt.scatter(df2[0], df2[1], color='g', alpha=0.5)
plt.grid()
plt.show()
```

```python
# make_moons 三日月形のデータ
from sklearn.datasets import make_moons
X, y = make_moons(
    random_state=3,
    noise=0.1,     # ノイズ,ばらつき具合
    n_samples=300
)

# df
df = pd.DataFrame(X)
df['target'] = y
df0 = df[df['target']==0]
df1 = df[df['target']==1]
# 散布図描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color='b', alpha=0.5)
plt.scatter(df1[0], df1[1], color='r', alpha=0.5)
plt.grid()
plt.show()
```

```python
# make_circles 二重円状のデータ
from sklearn.datasets import make_circles
X, y = make_circles(
    random_state=3,
    noise=0.05,    
    n_samples=300
)

# df
df = pd.DataFrame(X)
df['target'] = y
df0 = df[df['target']==0]
df1 = df[df['target']==1]
# 散布図描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color='b', alpha=0.5)
plt.scatter(df1[0], df1[1], color='r', alpha=0.5)
plt.grid()
plt.show()
```

```python
# make_gaussian_quantiles 同心円状のデータ
from sklearn.datasets import make_gaussian_quantiles
X, y = make_gaussian_quantiles(
    random_state=3,
    n_features=2, # 特徴量の数    2次元
    n_classes=3,  # グループの数  3つのグループ
    n_samples=300
)

# df
df = pd.DataFrame(X)
df['target'] = y
df0 = df[df['target']==0]
df1 = df[df['target']==1]
df2 = df[df['target']==2]
# 散布図描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color='b', alpha=0.5)
plt.scatter(df1[0], df1[1], color='r', alpha=0.5)
plt.scatter(df2[0], df2[1], color='g', alpha=0.5)
plt.grid()
plt.show()
```

```python
# make_regression 回帰のデータ
from sklearn.datasets import make_regression
X, y = make_regression(
    random_state=3,
    n_features=1, # 特徴量の数 
    noise=0,     # ノイズ
    bias=0,     # y切片
    n_samples=300
)

# df
df = pd.DataFrame(X)
# 散布図描画
plt.figure(figsize=(5,5))
plt.scatter(df[0], y, color='b', alpha=0.5)
plt.grid()
plt.show()
```
