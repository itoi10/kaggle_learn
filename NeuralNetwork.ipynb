```python
import numpy as np
import matplotlib.pyplot as plt
```

### 活性化関数
 - シグモイド
 - ReLU

```python
# シグモイド関数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU関数
def relu(x):
    return np.maximum(0, x)
```

```python
# sigmoid
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.title('Sigmoid')
plt.show()

# ReLU
x = np.arange(-5, 5, 0.1)
y = relu(x)

plt.plot(x, y)
plt.ylim(-1, 5.5)
plt.title('ReLU')
plt.show()
```

### softmax関数

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

## 3層ニューラルネットワーク

```python
class NeuralNetwork:

    def __init__(self):
        """ネットワークの重み(W),バイアス(b)を初期化"""

        self.network = {}

        self.network['W1'] = np.array([ [0.1, 0.3, 0.5],
                                        [0.2, 0.4, 0.6]])
        self.network['b1'] = np.array([0.1, 0.2, 0.3])

        self.network['W2'] = np.array([ [0.1, 0.4],
                                        [0.2, 0.5],
                                        [0.3, 0.6] ])
        self.network['b2'] = np.array([0.1, 0.2])

        self.network['W3'] = np.array([ [0.1, 0.3],
                                        [0.2, 0.4]])
        self.network['b3'] = np.array([0.1, 0.2])


    def forward(self, x):
        """順伝搬"""

        W1, W2, W3 = self.network['W1'], self.network['W2'], self.network['W3']
        b1, b2, b3 = self.network['b1'], self.network['b2'], self.network['b3']

        ## 1層目
        # A = XW + B
        a1 = np.dot(x, W1) + b1
        # 活性化関数
        z1 = sigmoid(a1)

        ## 2層目
        a2 = np.dot(z1, W2) + b2
        z2 = sigmoid(a2)

        ## 3層目
        a3 = np.dot(z2, W3) + b3
        y = softmax(a3)

        return y
```

#### １層目の行列式

$$
A = XW + B
$$

$$
\begin{bmatrix} 
a1 & a2 & a3\\
\end{bmatrix}
=
\begin{bmatrix} 
x1 & x2 \\ 
\end{bmatrix}

\begin{bmatrix} 
w11 & w21 & w31 \\ 
w12 & w22 & w32 \\ 
\end{bmatrix}
+
\begin{bmatrix} 
b1 & b2 & b3 \\ 
\end{bmatrix}
$$


```python
nn = NeuralNetwork()
x = np.array([1.0, 0.5])
y = nn.forward(x)
print(y)
print(y[0] + y[1])
```
